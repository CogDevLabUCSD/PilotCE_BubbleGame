```r
rm(list = ls())
library(plyr)
#library(agricolae)
# Change this to your filepath

filepath = '/Users/kjenson/Desktop/Projects/pilotCE_bubbleGame/channels/cleanData/'

setwd(filepath)

EEG = read.csv('CE_cleaned_clean_erp_FULL.csv')
```

```r
## Run rmANOVA’S on just unexpected block data using for the CE channel data

##### Again, the very first thing we'll do is narrow it down to 3 electrodes. 
# This can be removed easily in the future as there will be subsetting later.

EEG_justChannels = subset(EEG, channel_cluster == 'Fz' | channel_cluster == 'Cz' | channel_cluster == 'Oz')

## We only want to use the uncertain blocks, so we'll filter using the miniblock numbers
EEG_justUncertain = droplevels(subset(EEG_justChannels, miniBlock == 'Block: 2' | miniBlock == 'Block: 2b' | miniBlock == 'Block: 4' | miniBlock == 'Block: 4b'))

# We'll then create within-subject averages for each trial type
EEG_subjectConditionAverages = ddply(EEG_justUncertain, c('snum', 'rtype', 'continge', 'who', 'channel_cluster'), 'summarize', mean400ms = mean(c390_410), sd400ms = sd(c390_410), mean520ms = mean(c500_540), sd520ms = sd(c500_540))
```

```r
###### We'll first look at channel Fz
EEG_channelFz = subset(EEG_subjectConditionAverages, channel_cluster == 'Fz')

# now we can run our rmANOVA
# we'll just include rtype, condtion, and who as within subject factors

# first, we'll look at mean400ms
channelFz_400ms_aov = with(subset(EEG_channelFz ), aov(mean400ms~rtype*continge*who + Error(snum/(rtype*continge*who))))
print('Effect of Reward, Continge, and Who: Fz @ 400ms')
summary(channelFz_400ms_aov)

# next, we'll look at 520ms
channelFz_520ms_aov = with(subset(EEG_channelFz ), aov(mean520ms~rtype*continge*who + Error(snum/(rtype*continge*who))))
print('Effect of Reward, Continge, and Who: Fz @ 520ms')
summary(channelFz_520ms_aov)



###### Next we'll look at channel Cz
EEG_channelCz = subset(EEG_subjectConditionAverages, channel_cluster == 'Cz')

# now we can run our rmANOVA
# we'll just include rtype, condtion, and who as within subject factors

# first, we'll look at mean400ms
channelCz_400ms_aov = with(subset(EEG_channelCz ), aov(mean400ms~rtype*continge*who + Error(snum/(rtype*continge*who))))
print('Effect of Reward, Continge, and Who: Cz @ 400ms')
summary(channelCz_400ms_aov)

# next, we'll look at 520ms
channelCz_520ms_aov = with(subset(EEG_channelCz ), aov(mean520ms~rtype*continge*who + Error(snum/(rtype*continge*who))))
print('Effect of Reward, Continge, and Who: Cz @ 520ms')
summary(channelCz_520ms_aov)





###### Next we'll look at channel Oz
EEG_channelOz = subset(EEG_subjectConditionAverages, channel_cluster == 'Oz')

# now we can run our rmANOVA
# we'll just include rtype, condtion, and who as within subject factors

# first, we'll look at mean400ms
channelOz_400ms_aov = with(subset(EEG_channelOz ), aov(mean400ms~rtype*continge*who + Error(snum/(rtype*continge*who))))
print('Effect of Reward, Continge, and Who: Oz @ 400ms')
summary(channelOz_400ms_aov)

# next, we'll look at 520ms
channelOz_520ms_aov = with(subset(EEG_channelOz ), aov(mean520ms~rtype*continge*who + Error(snum/(rtype*continge*who))))
print('Effect of Reward, Continge, and Who: Oz @ 520ms')
summary(channelOz_520ms_aov)
```

```r
## This script is used to clean the data and might be uneccesary in this context
rm(list = ls())
library(plyr)

# Change this to your filepath
filepath = '/Users/kjenson/Desktop/Projects/pilotCE_bubbleGame/channels/rawdata'
setwd(filepath)
EEG = read.csv('CE_clean_erp.csv')
```

```r
# First remove NA's and all trials which are not one of the 4 main types (i.e. normal HR, reverse HR, normal LR, and reverse LR)
EEG_noNA = na.omit(EEG)

EEG_noBadTrials = droplevels(subset(EEG_noNA, compos == 'normal_high reward' | compos == 'normal_low reward' | compos == 'reverse_high reward' | compos == 'reverse_low reward'))

## use this to check for effectiveness
levels(EEG_noBadTrials$compos)
```

```r
## Add dyadID/pairID (i.e. was it dyad c12, e29, etc.)
EEG_noBadTrials$pairID = as.factor(substr(EEG_noBadTrials$snum, 1, 3))


## Add side info
EEG_noBadTrials$side = NA

EEG_sideIncluded = data.frame(matrix(ncol = length(colnames(EEG_noBadTrials)), nrow = 0))
colnames(EEG_sideIncluded) <- colnames(EEG_noBadTrials)

justRight = subset(EEG_noBadTrials, type == 'mom1' | type == 'baby1')
justRight$side = 'momRight'

justLeft = subset(EEG_noBadTrials, type == 'mom2' | type == 'baby2')
justLeft$side = 'momLeft'

EEG_sideIncluded = rbind(justLeft, justRight)


## Check to make sure it worked (i.e. that momRight and momLeft are the only two factor levels for side)
levels(as.factor(EEG_sideIncluded$side))
```

```r
# Lastly, get rid of everybody not in the study set
EEG_justGoodSubs = droplevels(subset(EEG_sideIncluded, snum == 'c13b' | snum == 'c13m' | snum == 'c14b' | snum == 'c14m' | snum == 'c16b' | snum == 'c17b' | snum == 'c18m' | snum == 'c19b' | snum == 'c19m' | snum == 'c20b' | snum == 'c20m' | snum == 'c21b' | snum == 'c21m' | snum == 'c22b' | snum == 'c22m' | snum == 'c23m' | snum == 'c24m' | snum == 'c25b' | snum == 'c25m' | snum == 'c26b' | snum == 'c28b' | snum == 'c28m' | snum == 'c29b' | snum == 'c29m' | snum == 'c30b' | snum == 'c30m' | snum == 'e11b' | snum == 'e11m' | snum == 'e13b' | snum == 'e13m' | snum == 'e15b' | snum == 'e15m' | snum == 'e18b' | snum == 'e18m' | snum == 'e01b' | snum == 'e01m' | snum == 'e20b' | snum == 'e20m' | snum == 'e22b' | snum == 'e22m' | snum == 'e23b' | snum == 'e23m' | snum == 'e24b' | snum == 'e24m' | snum == 'e26b' | snum == 'e26m' | snum == 'e27b' | snum == 'e27m' | snum == 'e28b' | snum == 'e28m' | snum == 'e29b' | snum == 'e29m' | snum == 'e30b' | snum == 'e30m' | snum == 'e31b' | snum == 'e31m' | snum == 'e04b' | snum == 'e04m' | snum == 'e09b' | snum == 'e09m')) 

# Only those in the study set should remain
levels(droplevels(as.factor(EEG_justGoodSubs$snum)))
```
